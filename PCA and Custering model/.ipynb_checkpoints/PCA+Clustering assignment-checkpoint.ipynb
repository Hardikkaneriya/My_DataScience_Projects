## Hopkins analysis

### Hopkins method to check whether we can built the cluster or not 

```python
from sklearn.neighbors import NearestNeighbors
from random import sample
from numpy.random import uniform
import numpy as np
from math import isnan
 
def hopkins(X):
    d = X.shape[1]
    #d = len(vars) # columns
    n = len(X) # rows
    m = int(0.1 * n) 
    nbrs = NearestNeighbors(n_neighbors=1).fit(X.values)
 
    rand_X = sample(range(0, n, 1), m)
 
    ujd = []
    wjd = []
    for j in range(0, m):
        u_dist, _ = nbrs.kneighbors(uniform(np.amin(X,axis=0),np.amax(X,axis=0),d).reshape(1, -1), 2, return_distance=True)
        ujd.append(u_dist[0][1])
        w_dist, _ = nbrs.kneighbors(X.iloc[rand_X[j]].values.reshape(1, -1), 2, return_distance=True)
        wjd.append(w_dist[0][1])
 
    H = sum(ujd) / (sum(ujd) + sum(wjd))
    if isnan(H):
        print(ujd, wjd)
        H = 0
 
    return H
```

```python
hopkins(country_data_pca)
```

##### Above result shows that data is well organized to form the clusters

##  K-means Clustering method

  ### silhouette_score

```python
from sklearn.metrics import silhouette_score
sse_ = []
for k in range(2,15):
    kmeans = KMeans(n_clusters=k).fit(country_data_pca)
    sse_.append([k, silhouette_score(country_data_pca, kmeans.labels_)])
```

```python
plt.plot(pd.DataFrame(sse_)[0], pd.DataFrame(sse_)[1])
```

##### from the graph it is clear that K= 3 needs to choose

# sum of squared distances analysis

```python
# sum of squared distances
ssd = []
for num_clusters in list(range(1,21)):
    model_clus = KMeans(n_clusters = num_clusters, max_iter=60)
    model_clus.fit(country_data_pca)
    ssd.append(model_clus.inertia_)

plt.plot(ssd)
```

###### above elbow curve suggest K=3

## Form the K-means cluster with K=3

```python
# Kmeans with K=3
model_clus3 = KMeans(n_clusters = 3, max_iter=60 ,random_state=50)
model_clus3.fit(country_data_pca)
```

```python
# concatenate the cluster Id to each rows of country_data_pca
country_data_pca.index = pd.RangeIndex(len(country_data_pca.index))
country_data_pca_km = pd.concat([country_data_pca, pd.Series(model_clus3.labels_)], axis=1)
country_data_pca_km.columns = ['PC1', 'PC2','PC3','PC4','PC5','ClusterID']
```
